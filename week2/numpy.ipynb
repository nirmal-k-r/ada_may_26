```python
marks=[90,80,75,60]

total=0
for mark in marks:
    total+=mark

average=total/len(marks)
print("Average mark:", average)
```

```python
import numpy as np

marks=np.array([90,80,75,60])
average=np.mean(marks)
print("Average mark:", average)

sum=np.sum(marks)
min=np.min(marks)
max=np.max(marks)

std=np.std(marks)

print(f'Sum {sum}, Min {min}, Max {max}, Std {std}')
```
