```python
#plot rarity class distribution
sns.countplot(x='rarity_class', data=df)
plt.title('Rarity Class Distribution')
plt.xlabel('Rarity Class')
plt.ylabel('Count') 
```

```python
#pie chart of rarity class distribution
plt.figure(figsize=(8,8))
plt.pie(df['rarity_class'].value_counts(), labels=df['rarity_class'].unique(), autopct='%1.1f%%')
plt.title('Rarity Class Distribution')
plt.show()
```
